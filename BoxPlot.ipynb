```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
oil_prices = pd.read_excel("February_oil_prices.xlsx")
print(oil_prices)
```

```python
brent = oil_prices["Brent"]
wti = oil_prices["WTI"]
labels = ["Brent", "WTI"]
plt.boxplot([brent, wti], labels=labels)
plt.show()
```

```python
brent = oil_prices["Brent"]
wti = oil_prices["WTI"]
labels = ["Brent", "WTI"]
plt.title("February oil prices fro Brent and WTI")
plt.xlabel("Oil Blend")
plt.ylabel("Price per barrel in USD")
plt.boxplot([brent, wti], labels=labels, patch_artist=True, 
            boxprops=dict(facecolor="#aaae02", linewidth=3),
            whiskerprops=dict(linewidth=3),
            medianprops=dict(color="#ffaaff", linewidth=5)
)
plt.show()
```

```python
brent = oil_prices["Brent"]
wti = oil_prices["WTI"]
labels = ["Brent", "WTI"]
plt.title("February oil prices fro Brent and WTI")
plt.xlabel("Oil Blend")
plt.ylabel("Price per barrel in USD")
plt.boxplot([brent, wti], labels=labels, patch_artist=True, notch="True",
            boxprops=dict(facecolor="#aaae02", linewidth=3),
            whiskerprops=dict(linewidth=3),
            medianprops=dict(color="#ffaaff", linewidth=5)
)
plt.show()
```

```python
brent = oil_prices["Brent"]
wti = oil_prices["WTI"]
labels = ["Brent", "WTI"]
plt.title("February oil prices fro Brent and WTI")
plt.xlabel("Oil Blend")
plt.ylabel("Price per barrel in USD")
plt.boxplot([brent, wti], labels=labels, patch_artist=True, vert = 0,
            boxprops=dict(facecolor="#aaae02", linewidth=3),
            whiskerprops=dict(linewidth=3),
            medianprops=dict(color="#ffaaff", linewidth=5)
)
plt.show()
```
